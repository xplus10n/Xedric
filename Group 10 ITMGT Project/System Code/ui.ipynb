```python
import import_ipynb
import pygame
from settings import *
from random import choice

class UI:
    def __init__(self, monster, get_input):
        self.display_surface = pygame.display.get_surface()
        self.font = pygame.font.Font(None, 30)
        self.left = WINDOW_WIDTH/2 - 100
        self.top = WINDOW_HEIGHT/2 + 50
        self.monster = monster
        self.get_input = get_input
        
        # control
        self.general_options = ['Attack', 'Defense', 'Charge', 'Repair']
        self.general_index = {'col':0, 'row':0}
        self.attack_index = {'col':0, 'row':0}
        
        self.Ch1ATK_index = {'col':0, 'row':0}
        self.Ch2ATK_index = {'col':0, 'row':0}
        self.Ch3ATK_index = {'col':0, 'row':0}
        self.Ch4ATK_index = {'col':0, 'row':0}

        self.Dranken_index = {'col':0, 'row':0}        
        self.Sniper_index = {'col':0, 'row':0}
        self.Split_Dranken_index = {'col':0, 'row':0}
        self.Fireball_index = {'col':0, 'row':0}
        self.Shotgun_index = {'col':0, 'row':0}
        
        self.defense_index = {'col':0, 'row':0}
        self.state = 'general'
        self.rows, self.cols = 2,2
        
        # moves storage
        self.turn_moves = []
        self.targets = []

        # starting charges
        self.player1_charges = 1
        self.player2_charges = 1
        self.player3_charges = 1
        self.player4_charges = 1

        # displayed
        self.attacks = list(ATTACK_DATA.keys())
        self.Ch1ATK = list(ATTACK_DATA['1 Ch ATK'])
        self.Ch2ATK = list(ATTACK_DATA['2 Ch ATK'])
        self.Ch3ATK = list(ATTACK_DATA['3 Ch ATK'])
        self.Ch4ATK = list(ATTACK_DATA['4+ Ch ATK'])
        self.defenses = DEFENSE_DATA
        self.single_target = ['player 1', 'player 2', 'player 3', 'player 4']
        self.double_target = ['Not player 1', 'Not player 2', 'Not player 3', 'Not player4']

    def input(self):
        keys = pygame.key.get_just_pressed()

        # scrolling through menu
        if self.state == 'general':
            self.rows, self.cols = 2,2
            self.general_index['row'] = (self.general_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.general_index['col'] = (self.general_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                self.state = self.general_options[self.general_index['col'] + self.general_index['row'] * 2]
                

        elif self.state == 'Attack':
            self.rows, self.cols = 2,2
            self.attack_index['row'] = (self.attack_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.attack_index['col'] = (self.attack_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                Attack = self.attacks[self.attack_index['col'] + self.attack_index['row'] * 2]
                self.state = Attack

        elif self.state == '1 Ch ATK':
            self.rows, self.cols = 2,2
            self.Ch1ATK_index['row'] = (self.Ch1ATK_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Ch1ATK_index['col'] = (self.Ch1ATK_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                Ch1ATK = self.Ch1ATK[self.Ch1ATK_index['col'] + self.Ch1ATK_index['row'] * 2] 
                self.state = Ch1ATK

        elif self.state == 'Dranken':
            self.rows, self.cols = 2,2
            self.Dranken_index['row'] = (self.Dranken_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Dranken_index['col'] = (self.Dranken_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                single_target = self.single_target[self.Dranken_index['col'] + self.Dranken_index['row'] * 2]

                #experimental
                self.turn_moves.append('Dranken') 
                self.targets.append(single_target) 
                self.get_input(self.state, single_target)
                self.state = 'general'  

        elif self.state == 'Sniper':
            self.rows, self.cols = 2,2
            self.Sniper_index['row'] = (self.Sniper_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Sniper_index['col'] = (self.Sniper_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                single_target = self.single_target[self.Sniper_index['col'] + self.Sniper_index['row'] * 2]

                #experimental
                self.turn_moves.append('Sniper') 
                self.targets.append(single_target) 
                self.get_input(self.state, single_target)
                self.state = 'general'    

        elif self.state == 'Split Dranken':
            self.rows, self.cols = 2,2
            self.Split_Dranken_index['row'] = (self.Split_Dranken_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Split_Dranken_index['col'] = (self.Split_Dranken_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                double_target = self.double_target[self.Split_Dranken_index['col'] + self.Split_Dranken_index['row'] * 2]

                #experimental
                self.turn_moves.append('Split Dranken') 
                self.targets.append(double_target) 
                self.get_input(self.state, double_target)
                self.state = 'general'    
        
        elif self.state == '2 Ch ATK':
            self.rows, self.cols = 2,2
            self.Ch2ATK_index['row'] = (self.Ch2ATK_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Ch2ATK_index['col'] = (self.Ch2ATK_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                Ch2ATK = self.Ch2ATK[self.Ch2ATK_index['col'] + self.Ch2ATK_index['row'] * 2]

                #experimental
                self.turn_moves.append(Ch2ATK)
                
                self.get_input(self.state, Ch2ATK)
                self.state = 'general'
                self.targets.append('AOE') 

        elif self.state == '3 Ch ATK':
            self.rows, self.cols = 2,2
            self.Ch3ATK_index['row'] = (self.Ch3ATK_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Ch3ATK_index['col'] = (self.Ch3ATK_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                Ch3ATK = self.Ch3ATK[self.Ch3ATK_index['col'] + self.Ch3ATK_index['row'] * 2]

                #experimental                
                self.state = Ch3ATK

        elif self.state == 'Fireball':
            self.rows, self.cols = 2,2
            self.Fireball_index['row'] = (self.Fireball_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Fireball_index['col'] = (self.Fireball_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                double_target = self.double_target[self.Fireball_index['col'] + self.Fireball_index['row'] * 2]

                #experimental
                self.turn_moves.append('Fireball')
                self.targets.append(double_target) 
                self.get_input(self.state, double_target)
                self.state = 'general'    
        
        elif self.state == 'Shotgun':
            self.rows, self.cols = 2,2
            self.Shotgun_index['row'] = (self.Shotgun_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Shotgun_index['col'] = (self.Shotgun_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                double_target = self.double_target[self.Shotgun_index['col'] + self.Sniper_index['row'] * 2]

                #experimental
                self.turn_moves.append('Shotgun')
                self.targets.append(double_target) 
                self.get_input(self.state, double_target)
                self.state = 'general'    

        elif self.state == '4+ Ch ATK':
            self.rows, self.cols = 2,2
            self.Ch4ATK_index['row'] = (self.Ch4ATK_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.Ch4ATK_index['col'] = (self.Ch4ATK_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                Ch4ATK = self.Ch4ATK[self.Ch4ATK_index['col'] + self.Ch4ATK_index['row'] * 2]
                self.get_input(self.state, Ch4ATK)
                self.state = 'general'

                #experimental
                self.turn_moves.append(Ch4ATK) 
                self.targets.append('AOE') 
        
        elif self.state == 'Defense':
            self.rows, self.cols = 2,2
            self.defense_index['row'] = (self.defense_index['row'] + int(keys[pygame.K_DOWN]) - int(keys[pygame.K_UP])) % self.rows
            self.defense_index['col'] = (self.defense_index['col'] + int(keys[pygame.K_RIGHT]) - int(keys[pygame.K_LEFT])) % self.cols
    
            # selecting a key in menu
            if keys[pygame.K_SPACE]:
                Defense = self.defenses[self.defense_index['col'] + self.defense_index['row'] * 2]
                self.get_input(self.state, Defense)
                self.state = 'general'

                #experimental
                self.turn_moves.append(Defense)
                self.targets.append('AOE') 
        
        elif self.state == 'Charge':
            self.get_input('Charge', '+1')
            self.state = 'general'
            
            #experimental
            self.turn_moves.append('Charge')  
            self.targets.append('AOE') 
        
        elif self.state == 'Repair':
            self.get_input('Repair', 'Shield Repaired')
            self.state = 'general'                 

            #experimental
            self.turn_moves.append('Repair')   
            self.targets.append('AOE') 
        
        if keys[pygame.K_ESCAPE]:
            self.state = 'general'
            self.general_index = {'col':0, 'row':0}
            self.attack_index = {'col':0, 'row':0}
            self.Ch1ATK_index = {'col':0, 'row':0}
            self.Ch2ATK_index = {'col':0, 'row':0}
            self.Ch3ATK_index = {'col':0, 'row':0}
            self.Ch4ATK_index = {'col':0, 'row':0}
            self.defense_index = {'col':0, 'row':0}
            self.Dranken_index = {'col':0, 'row':0}        
            self.Sniper_index = {'col':0, 'row':0}
            self.Split_Dranken_index = {'col':0, 'row':0}
            self.Fireball_index = {'col':0, 'row':0}
            self.Shotgun_index = {'col':0, 'row':0}
           
    
    def quad_select(self, index, options):
        # bg
        rect = pygame.FRect(self.left + 40,self.top + 60,400,200)
        pygame.draw.rect(self.display_surface, COLORS['white'], rect, 0, 4)
        pygame.draw.rect(self.display_surface, COLORS['gray'], rect, 4, 4)

        # menu
        for col in range(self.cols):
            for row in range(self.rows):
                x = rect.left + rect.width / (self.cols * 2) + (rect.width / self.cols) * col
                y = rect.top + rect.height / (self.rows * 2) + (rect.height / self.rows) * row
                i = col + 2 * row
                color = COLORS['gray'] if col == index['col'] and row == index['row'] else COLORS['black']
                
                text_surf = self.font.render(options[i], True, color)
                text_rect = text_surf.get_frect(center = (x,y))
                self.display_surface.blit(text_surf, text_rect)

                

    def update(self):
        self.input()
    
    def draw(self):
        match self.state:
            case 'general': self.quad_select(self.general_index, self.general_options)
            case 'Attack': self.quad_select(self.attack_index, self.attacks) 
            case 'Defense': self.quad_select(self.defense_index, self.defenses) 
            case '1 Ch ATK': self.quad_select(self.Ch1ATK_index, self.Ch1ATK) 
            case '2 Ch ATK': self.quad_select(self.Ch2ATK_index, self.Ch2ATK) 
            case '3 Ch ATK': self.quad_select(self.Ch3ATK_index, self.Ch3ATK) 
            case '4+ Ch ATK': self.quad_select(self.Ch4ATK_index, self.Ch4ATK) 
            case 'Dranken' : self.quad_select(self.Dranken_index, self.single_target) 
            case 'Sniper' : self.quad_select(self.Sniper_index, self.single_target)
            case 'Split Dranken' : self.quad_select(self.Split_Dranken_index, self.double_target)
            case 'Fireball' : self.quad_select(self.Fireball_index, self.double_target)
            case 'Shotgun' : self.quad_select(self.Shotgun_index, self.double_target)
 
```
